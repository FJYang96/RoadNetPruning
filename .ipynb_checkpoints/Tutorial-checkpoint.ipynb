# Simple Example

```python
# Create a NetworkX graph
G = nx.DiGraph()
N, M = 15, 80
edges = np.random.randint(N, size=(M,2))
G.add_edges_from(edges)

# Sample the cost and capacity of the edges
capacities = np.random.uniform(1000, size=(M,))
costs = np.random.uniform(10, size=(M,))
for i, e in enumerate(G.edges):
    G.edges[e]['capacity'] = capacities[i]
    G.edges[e]['weight'] = costs[i]

# Sample a demand
origs, dests, rates = sample_demand(G, rate_range=(5,10))
demand = Demand(origs, dests, rates)
demand_backup = demand

# Solve the sparsification problem with rntools
simple = rntools.core.LPSparsifier(G, demand, lbda=0)
simple.solve()
print('Cost is', simple.compute_cost())
simple.nx_draw()
```

# Greedy Spanner

```python
def greedy_spanner(G, factor):
    # Add the nodes first
    sp = nx.MultiDiGraph()
    sp.graph = G.graph
    sp.add_nodes_from(G.nodes(data=True))

    # Loop through the edges in ascending order
    for s, t, d in sorted(G.edges(data=True), key=lambda x: x[2]['weight']):
        orig_dist = d['weight']
        sp_dist = np.inf
        try:
            sp_dist = nx.algorithms.shortest_path_length(sp, s,t, 'weight')
        except:
            pass
        if sp_dist > factor * orig_dist:
            sp.add_edge(s,t, **d)
    return sp
```

```python
G = nx.DiGraph()
N, M = 75, 700
edges = np.random.randint(N, size=(M,2))
G.add_edges_from(edges)
#weights = np.random.uniform(10, size=(M,))
weights = np.ones(M)
for i, e in enumerate(G.edges):
    G.edges[e]['weight'] = weights[i]
nx.draw(G)
```

```python
counter = 0
while counter < 50:
    kamppi_graph = rntools.io.load_graph('./data/Kamppi.graphml')
    rntools.utils.add_cost_and_capacity_metric(kamppi_graph)
    kamppi_origs, kamppi_dests, kamppi_rates = sample_demand(kamppi_graph, num_od_pairs=15,
                                                             rate_range=(150,250))
    kamppi_demand = Demand(kamppi_origs, kamppi_dests, kamppi_rates)
    kamppi_lp = rntools.core.LPSparsifier(kamppi_graph, kamppi_demand, lbda=0)
    if kamppi_lp.solve(): #kamppi_lp.ox_draw()
        u = kamppi_lp.solution / kamppi_lp.capacity
    else:
        u = np.zeros(238)
        counter -= 1
    counter += 1
    print(counter, end=',')
    usage += u
```

```python
def get_subgraph_from_solution(graph, sparsifier, thres=0):
    edge_usage = sparsifier.solution / sparsifier.capacity
    non_zero_inds = np.nonzero(edge_usage > thres)[0]
    edges = list(graph.edges)
    edge_list = []
    for i in non_zero_inds:
        edge_list.append(edges[i])
    subgraph = graph.edge_subgraph(edge_list).copy()
    return subgraph

def iterative_sparsification(graph, demand, lbda=0, num_iterations=5, thres=0.001):
    num_edges = np.zeros(num_iterations+1)
    num_edges[0] = len(graph.edges)
    print(num_edges[0])
    subgraph = graph.copy()
    for i in range(num_iterations):
        iterative_lp = rntools.core.LPSparsifier(subgraph, demand, lbda)
        iterative_lp.solve()
        subgraph = get_subgraph_from_solution(subgraph, iterative_lp, thres=thres)
        print(len(subgraph.edges))
        num_edges[i+1] = len(subgraph.edges)
    return num_edges
```
