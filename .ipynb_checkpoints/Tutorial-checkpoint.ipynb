# Simple Example

```python
# Create a NetworkX graph
G = nx.DiGraph()
N = 15
M = 80
edges = np.random.randint(N, size=(M,2))
G.add_edges_from(edges)

nx.draw_circular(G, with_labels=True)
```

```python
# Sample the cost and capacity of the edges
capacities = np.random.uniform(100, size=(M,))
costs = np.random.uniform(10, size=(M,))
for i, e in enumerate(G.edges):
    G.edges[e]['cost'] = costs[i]
    G.edges[e]['capacity'] = capacities[i]
```

```python
# Sample a demand
origs, dests, rates = sample_demand(G)
demand = Demand(origs, dests, rates)
demand_backup = demand
print(list(zip(origs, dests)))
```

```python
rn = rntools.core.RNTools(G, demand)
```

```python
rn.construct_LP(lbda=10, phi=17, num_samples=10)
```

```python
rn.nx_draw()
```

# Simple Road Network Example

```python
# Load the graph
graph = rntools.io.load_graph('./data/Kamppi.graphml')
rntools.utils.add_cost_and_capacity_metric(graph)
rn_origs, rn_dests, rn_rates = sample_demand(graph, num_od_pairs=30, rate_range=(0,0.005))
rn_demand = Demand(rn_origs, rn_dests, rn_rates)
```

```python
nx.is_strongly_connected(graph)
```

```python
kamppi = rntools.core.RNTools(graph, rn_demand)
kamppi.construct_LP(lbda=0, phi=0, num_samples=100)
```

```python
od = kamppi.ox_draw()
```
